```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("total_me_data.csv")
```

```python
# back_Test용만 뺀다.
back_test_df = df.query('Date == "2019-06-28 0:00" | Date == "2019-09-30 0:00" | Date == "2019-03-29 0:00" |  Date == "2018-12-28 0:00"')
```

# Model Training

```python
# bacK_test 용은 삭제
df.drop(back_test_df.isin(df).index, inplace=True)
```

```python
df.drop('MA',1)
```

```python
#features_treain, labels_train
train_set = df # df는 전체 train_set
features_train = winsorize(train_set.iloc[:, 2:8], 0.05)
labels_train = winsorize(train_set.iloc[:, 8:], 0.05)
```

```python
#Classifier 선언
clf = GaussianNB()
#fit 함수가 training sample을 이용한 학습을 진행하는 과정
clf.fit(features_train, labels_train)
```
