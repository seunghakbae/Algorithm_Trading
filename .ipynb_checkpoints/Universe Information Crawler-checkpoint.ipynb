```python
#0-1) Finance.Naver에서 주가정보를 크롤링: '종가'만 출력하도록 변경

def priceClose_naver(symbol, timeframe, count):
    
    #making url
    
    url_a = 'https://fchart.stock.naver.com/sise.nhn?requestType=0'
    url_insert =  url_a+'&symbol='+symbol+'&timeframe='+timeframe+'&count='+count

    #convert into bs(beautifulsoup) object
    
    price_raw = requests.get(url_insert)
    price_bs = bs4.BeautifulSoup(price_raw.text, 'lxml')
    price_list = price_bs.find_all('item')
    
    #empty sheets
    
    date_list = []; close_list = []

    #split the data into date/open/high/low/close/volume
    #close - 차트에서 끌어오기 때문에 수정종가로 자동반영
    
    for piece in price_list:
        temp = piece['data']
        dp = temp.split('|')
        date_list.append(dp[0]); close_list.append(dp[4])
    
    #dataframe으로 합치기
    
    dp_to_df = pd.DataFrame({'close': close_list}, index=date_list)
    dp_to_df = dp_to_df.astype(float)
    
    return dp_to_df
```

```python
#1-1) Wiseindex 사이트에서 섹터별 회사 정보를 가져오는 함수(w/o JSON)


def sectorList_wise(sector): #WICS기준, WI26기준 모두 사용 가능
    
    url = 'http://www.wiseindex.com/Index/GetIndexComponets?ceil_yn=0&dt=20190927&sec_cd=' + sector
    #비교 #http://www.wiseindex.com/Index/GetIndexComponets?ceil_yn=0&dt=20190927&sec_cd=G2520
    #왜 20190927이 고정되어있는데 작동이 되지??
    
    html = requests.get(url)
    bsObj = bs4.BeautifulSoup(html.text, 'lxml')
    data = str(bsObj.find_all('p'))
    
    temp = data.split("sector")[0] 
    temp = temp.split("list")[1]
    temp = temp[4:len(temp)-4]
    temp = temp.split("},{") #len(temp) #이제는 종목 개수를 반환한다.
    
    #empty sheet

    IDX_CD = []; IDX_NM_KOR = []; ALL_MKT_VAL = []; #업종 ID #업종 이름 #업종의 총 시장가치
    CMP_CD = []; CMP_KOR = []; MKT_VAL = []; #종목코드 #종목이름 #시장가치
    WGT = []; S_WGT = []; CAL_WGT = [] #시장 비중 #시장 비중(누적=1 되도록 오름차순이 기본) #합=1 인듯
    SEC_CD = []; SEC_NM_KOR = []; #WICS기준 정보: 업종 대분류 코드, 업종 이름
    SEQ = []; TOP60 = []; APT_SHR_CNT = [] #업종 내 순서(비중 순) #??? #???
    
    
    for i in range(0, len(temp)): 

        IDX_CD.append(temp[i].split(',')[0].split(":")[1].replace('"', ''))
        IDX_NM_KOR.append(temp[i].split(',')[1].split(":")[1].replace('"', ''))
        ALL_MKT_VAL.append(temp[i].split(',')[2].split(":")[1].replace('"', ''))
        CMP_CD.append(temp[i].split(',')[3].split(":")[1].replace('"', ''))
        CMP_KOR.append(temp[i].split(',')[4].split(":")[1].replace('"', ''))
        MKT_VAL.append(temp[i].split(',')[5].split(":")[1])
        WGT.append(temp[i].split(',')[6].split(":")[1])
        S_WGT.append(temp[i].split(',')[7].split(":")[1])
        CAL_WGT.append(temp[i].split(',')[8].split(":")[1])
        SEC_CD.append(temp[i].split(',')[9].split(":")[1].replace('"', ''))
        SEC_NM_KOR.append(temp[i].split(',')[10].split(":")[1].replace('"', ''))
        SEQ.append(temp[i].split(',')[11].split(":")[1])
        TOP60.append(temp[i].split(',')[12].split(":")[1])
        APT_SHR_CNT.append(temp[i].split(',')[13].split(":")[1])
    
    
    total_df = pd.DataFrame({"IDX_CD": IDX_CD, "IDX_NM_KOR": IDX_NM_KOR, "종목코드": CMP_CD, "종목명": CMP_KOR, "MKT_VAL": MKT_VAL,
                     "WGT": WGT, "S_WGT": S_WGT, "CAL_WGT": CAL_WGT, "SEC_CD": SEC_CD, "SEC_NM_KOR": SEC_NM_KOR,
                     "SEQ": SEQ, "TOP60": TOP60, "APT_SHR_CNT": APT_SHR_CNT})
    
    
    return total_df
```

```python
#1-2) JSON Parsing Practice
#1-2) Sector Valuation Information

def sectorValuation_wise(sector, from_dt, end_dt, term): #sector: WICS (WI26은 안해봄), from_dt, to_df = '00000000' 형식
    
    #term: 1 - 일간, 2 -주간, 3 - 월간, 4 - 분기
    
    ##라이브러리##
    #from urllib.request import urlopen #url사용
    #from pandas.io.json import json_normalize #JSON 형식 자동 dataframe 변환
    #from datetime import datetime #JSON 날짜 변환
    #import requests #??? - 필요없을 수도
    #import json #JSON Parsing을 위해 필요
    #import pandas as pd #Series, DataFrame Handling
    #import math #math.ceil -> 올림
    #import time #time.sleep -> 잦은 request로 인한 error 방지용
    #import re #optional #아래 markdown 날짜 변환의 두 번째 예시를 참조

    #url에 있었던 날짜 형식으로 변환
    k = from_dt[0:4] + '-' + from_dt[4:6] + '-' + from_dt[6:] 
    q = end_dt[0:4] + '-' + end_dt[4:6] + '-' + end_dt[6:]

    #1단계: 1페이지 먼저 로딩
    url_i ='http://www.wiseindex.com/DataCenter/GridData?currentPage=1&endDT='+q+'&fromDT='+k+'&index_ids='+sector+'&isEnd=1&itemType=3&perPage=100&term='+str(term)
    rqt = urllib.request.urlopen(url_i) #request
    datat = json.loads(rqt.read().decode()) #JSON 파싱 후
    df1 = json_normalize(datat) #1페이지 -> to_dataframe

    numer = datat[0]['TOT_ROW'] #총 Row의 갯수를 반환
    denom = 100 #default = 페이지당 100개 불러오기 (url)
    pg_cnt = math.ceil(numer/denom) #총 페이지수 계산

    
    #2단계: 2페이지부터 마지막 페이지까지 데이터프레임화 + 아래로 합치기
    for i in range(2, pg_cnt+1):
    
        try:
            print(i, 'processing')
            time.sleep(1) #1초에 한 번씩
        
            try: #페이지 i를 문자열화해서 받음. k, q, sector는 이미 함수 input으로 들어옴
                url_i2 = 'http://www.wiseindex.com/DataCenter/GridData?currentPage='+str(i)+'&endDT='+q+'&fromDT='+k+'&index_ids='+sector+'&isEnd=1&itemType=3&perPage=100&term='+str(term)
                rq2 = urllib.request.urlopen(url_i2)
                data2 = json.loads(rq2.read().decode())
                df2 = json_normalize(data2)
            
            except requests.exceptions.Timeout:
                time.sleep(60) #request exceptions timeout에 대해 60초 대기 후 다시 작업 재개 (위의 try와 작업 내용이 같음)
    
                url_i2 = 'http://www.wiseindex.com/DataCenter/GridData?currentPage='+str(i)+'&endDT='+q+'&fromDT='+k+'&index_ids='+sector+'&isEnd=1&itemType=3&perPage=100&term='+str(term)
                rq2 = urllib.request.urlopen(url_i2)
                data2 = json.loads(rq2.read().decode())
                df2 = json_normalize(data2)
            
            if i == 2: #2페이지는 1페이지와 합치고
                total = pd.concat([df1, df2])
    
            else: #3페이지 부터는 total(1+2 Page)과 합치기
                total = pd.concat([total, df2])
    
        except ValueError:
            continue
            
        except KeyError:
            continue
    
    #3단계: JSON 날짜변환
    for i in range(0, numer):
        
        d = total['TRD_DT'].iloc[i][6:-2]
        d = int(d[:10])
        d = datetime.fromtimestamp(d).strftime('%Y-%m-%d %I:%M:%S %p')
        total['TRD_DT'].iloc[i] = d
    
    #4단계: 결과 반환
    
    return total 

        
```

```python
#1-3) 네이버 금융에서 섹터 소분류별 URL 번호를 불러오는 함수

def sectorList_naver(a): 
    
    #1단계: 데이터 수집
   
    url = 'https://finance.naver.com/sise/sise_group.nhn?type=upjong' #참조사이트1
    rq = requests.get(url) #서버와의 통신
    soup = bs4.BeautifulSoup(urlopen(url), "html.parser") #통신 결과를 html.parser를 이용해 bs4 객체로 변환
    
    #2단계: 빈 리스트 생성
   
    sector_no = []
    sector_name = []

    #3단계: 반복문 처리를 통해 1)섹터 정보 링크 및 2)섹터 이름을 추출
    
    for link in soup.find_all('a'):
       
        if "/sise/sise_group_detail.nhn?type=upjong&no=" in link.get('href'): #섹터 분류 링크의 규칙 활용
            sector_no.append(link.get('href')[43:]) #링크의 번호를 storage에 저장
            sector_name.append(link.text) #링크의 섹터 이름을 sector_name에 저장
        
        else:
            pass
        
    #4단계: 데이터프레임 생성, 결과 표출
    
    frame = pd.DataFrame({'sector_no': sector_no, 'section': sector_name})
    
    return frame
```

```python
#앞에서 정의한 데이터의 내용을 바탕으로
#실제 소분류된 내용을 끌어오는 함수

def sectorData_naver(sector_no):
    
    url = 'https://finance.naver.com/sise/sise_group_detail.nhn?type=upjong&no='+str(sector_no)
    rq = requests.get(url)
    bsO = bs4.BeautifulSoup(rq.text, 'lxml')    

    title = bsO.find('title').text.split(':')[0]

    table = bsO.find_all('div', attrs={'class': 'box_type_l'})
    table = table[1].find_all('tr')
    table = table[2:len(table)-2]

    firm_code = []
    firm_name = []
    
    for i in table:
        firm_code.append(i.find('a').get('href').split('=')[1])
        firm_name.append(i.find('a').text)
        
    frame = pd.DataFrame({'종목코드': firm_code, '종목명': firm_name, '소분류': title})

    return frame
```
