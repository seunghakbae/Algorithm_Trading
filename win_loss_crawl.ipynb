```python
real_rate_list = list()
real_rate_list.append(rate_list[1])
real_rate_list.append(rate_list[2])
real_rate_list.append(rate_list[3])
real_rate_list.append(rate_list[5])
```

```python
real_rate_list
```

```python
test_table = rate_table[real_rate_list]
```

```python
len(test_table)
```

```python
test_table_copy = test_table.copy()
```

```python
for i in range(len(real_rate_list)):
    for j in range(len(test_table_copy)-1,0, -1):
        value = ((test_table[real_rate_list[i]][j-1] - test_table[real_rate_list[i]][j]) / test_table[real_rate_list[i]][j-1]) * 100
        test_table_copy[real_rate_list[i]][j-1] = value
```

```python
test_table_copy.insert(loc=0, column='Date', value=table['Date'])
```

```python
test_table_copy.insert(loc=5, column='total_revenue', value = table['매출액 Growth'])
```

```python
test_table_copy.insert(loc=6, column='total_profit', value = table['영업이익 Growth'])
```

```python
test_table_copy.drop(index=len(test_table_copy)-1, inplace=True)
```

```python
test_table_copy.rename(columns={"배당수익률(FY0)" : "Divedend yield ratio"}, inplace=True)
```
