```python
#데이터 핸들링, 기타
import numpy as np
import pandas as pd
import re
import math
import time
from datetime import datetime #JSON 날짜 변환

#시각화
import matplotlib.pylab as plt
import matplotlib.font_manager as fm
import seaborn as sns

#크롤링 및 JSON 핸들링
from urllib.request import urlopen
from pandas.io.json import json_normalize
import requests
import bs4
import json
```

```python
# Valuation Ratio 데이터 가지고 오는 url
# fromDT = 시작 날짜
# endDT = 끝 날짜
# perPage = 페이지당 가지고 오는 데이터 개수
# term = 일간/주간/월간/분기 를 가리키는 변수, 분기는 4
url_valuation_ratio = "http://www.wiseindex.com/DataCenter/GridData?currentPage=1&endDT=2019-11-06&fromDT=2010-05-07&index_ids=WGI0002T&isEnd=1&itemType=3&perPage=100&term=4"
```

```python
rqt = urlopen(url_valuation_ratio) #request
jason_data = json.loads(rqt.read().decode()) #JSON 파싱 후
value_ratio_df = json_normalize(jason_data) #1페이지 -> to_dataframe
```

```python
value_ratio_df # valuationr ratio 정보 담은 dataframe
```

```python
# Revenue_growth = 매출액 Growth
# Profit_growth =  영업이익 Growth
# Divedend_yield_ratio = 배당수익률(FY0)

title = {"IDX1_VAL1": "EPS",
"IDX1_VAL2": "BPS",
"IDX1_VAL3": "SPS",
"IDX1_VAL4": "PER",
"IDX1_VAL5": "PBR",
"IDX1_VAL6": "PSR",
"IDX1_VAL7": "EV/EVITDA",
"IDX1_VAL8": "Revenue_growth",
"IDX1_VAL9": "Profit_growth",
"IDX1_VAL10": "Divedend_yield_ratio",
"TRD_DT" : "Date"
}
```

```python
value_ratio_df.rename(columns=title, inplace=True) # value_ratio_df column 이름 바꾸기
```

```python
# Date 형식 바꾸기
for i in range(0, len(value_ratio_df)):

    d = value_ratio_df['Date'].iloc[i][6:-2] # /Date()/ 삭제하기
    d = int(d[:10]) # date 날짜 값 숫자로 바꾸기
    d = datetime.fromtimestamp(d).strftime('%Y-%m-%d %I:%M:%S %p') # 보기 쉬운 형식으로 변환
    value_ratio_df['Date'].iloc[i] = d
```

```python
# Date column 위치 앞 쪽으로 옮기기.
new_value_ratio_df_columns = list(value_ratio_df.columns) # value_ratio_df의 column 리스트로 바꾸기.
new_value_ratio_df_columns.pop() # Date 삭제
new_value_ratio_df_columns.insert(0, 'Date') # Date를 맨 앞에 넣기.
new_value_ratio_df_columns.remove('ROW_IDX') # ROW_IDX 삭제
new_value_ratio_df_columns.remove('TOT_ROW') # TOT_ROW 삭제
print(new_value_ratio_df_columns)
```

```python
# Date column 위치 앞 쪽으로 옮기기.
value_ratio_df = value_ratio_df.reindex(columns= new_value_ratio_df_columns)
```

```python
index_value_df.rename(columns=title, inplace=True) # index_value_df column명 바꾸기.
```

```python
# Date 형식 바꾸기
for i in range(0, len(value_ratio_df)):

    d = index_value_df['Date'].iloc[i][6:-2] # /Date()/ 삭제하기
    d = int(d[:10]) # date 날짜 값 숫자로 바꾸기
    d = datetime.fromtimestamp(d).strftime('%Y-%m-%d %I:%M:%S %p') # 보기 쉬운 형식으로 변환
    index_value_df['Date'].iloc[i] = d
```

```python
# Date column 위치 앞 쪽으로 옮기기.
index_value_df_columns = list(index_value_df.columns) # value_ratio_df의 column 리스트로 바꾸기.
index_value_df_columns.pop() # Date 삭제
index_value_df_columns.insert(0, 'Date') # Date를 맨 앞에 넣기.
index_value_df_columns.remove('ROW_IDX') # ROW_IDX 삭제
index_value_df_columns.remove('TOT_ROW') # TOT_ROW 삭제
print(index_value_df_columns)
```

```python
# Date column 위치 앞 쪽으로 옮기기.
index_value_df = index_value_df.reindex(columns= index_value_df_columns)
```

```python
# index_value_df랑 value_ratio_df Date기준으로 합치기.
df_temp = pd.merge(index_value_df, value_ratio_df, on='Date')
df_temp
```

```python
# 필요없는 column값 없애기.
temp = list(df_temp.columns)
final_df_columns = list() 
final_df_columns.append(temp[0]) # Date 넣기.
final_df_columns.append(temp[1]) # Index_Value 넣기
final_df_columns.append(temp[5]) # 배당수익률 넣기
final_df_columns.append(temp[6]) # BPS 넣기
final_df_columns.append(temp[8]) # PER 넣기
final_df_columns.append(temp[4]) # EPS 넣기
final_df_columns.append(temp[-1]) # Reveneue_growth 넣기
final_df_columns.append(temp[-2]) # Profit_growth 넣기
```

```python
final_df = df_temp[final_df_columns] # fianl_df_columns의 column만 가져오기.
```

```python
final_df # 증가분 제외한 final_df
```

```python
final_df_copy = final_df.copy() # win_loss값을 구하기 위한 copy
final_df_copy['W/L'] = range(len(final_df_copy)) # dummy값 넣기.
```

```python
# WIN_LOSS 구하기.

for i in range(len(final_df_copy)-1, 0, -1):
    if final_df['Index_value'][i-1] > final_df['Index_value'][i]:
        final_df_copy['W/L'].iloc[i-1] = 'W'
    else:
        final_df_copy['W/L'].iloc[i-1] = 'L'
```

```python
temp = list(final_df_copy.columns)
rate_column_need = list()
rate_column_need.append(temp[2]) # Divedend_yield_ratio 넣기
rate_column_need.append(temp[3]) # BPS 넣기
rate_column_need.append(temp[4]) # PER 넣기
rate_column_need.append(temp[5]) # EPS 넣기
```

```python
# 증감 분 구하기.
final_df_copy_rate = final_df_copy[rate_column_need]
```

```python
# 증감 분 copy 만들기.
final_df_copy_rate_copy = final_df_copy_rate.copy()
```

```python
# 증감 분 계산하기.
for i in range(len(rate_column_need)):
    for j in range(len(final_df_copy_rate)-1,0, -1):
        value = ((final_df_copy_rate[rate_column_need[i]][j-1] - final_df_copy_rate[rate_column_need[i]][j]) / final_df_copy_rate[rate_column_need[i]][j-1]) * 100
        final_df_copy_rate_copy[rate_column_need[i]][j-1] = value
```

```python
final_df_copy_rate = final_df_copy_rate_copy
del final_df_copy_rate_copy #필요없는 copy삭제
```

```python
final_df_copy_rate.insert(loc=0, column='Date', value=final_df['Date'])
final_df_copy_rate['Profit_growth'] = final_df['Profit_growth']
final_df_copy_rate['Revenue_growth'] =final_df['Revenue_growth']
final_df_copy_rate['W/L'] =final_df_copy['W/L']
```

```python
# 마지막 row 없애기.
final_df_copy_rate.drop(index=len(final_df_copy_rate)-1, inplace=True)
```

```python
# 이름 바꿔주기.
final_df_copy_rate.rename(columns={'Divedend_yield_ratio' : 'Divedend_yield_ratio_growth', 'BPS':'BPS_growth', 'PER' : 'PER_growth', 'EPS' : 'EPS_growth'}, inplace=True)
```

```python
final_df = final_df_copy_rate
del final_df_copy_rate
```

```python
# csv 파일로 만들기.
final_df.to_csv("final.csv")
```
